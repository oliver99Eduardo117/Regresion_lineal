```python
import statsmodels.api as sm           #Datos estadísticos
import statsmodels.formula.api as smf  #fórmulas de estilo R
import pandas as pd
url="https://raw.githubusercontent.com/oliver99Eduardo117/Regresion_lineal/main/Regresion_lineal.csv" #Liga del archivos csv 
df= pd.read_csv(url)
print(df.head())
```

```python
mod = smf.ols('y ~ x1 + x2', data=df).fit()  #Ajustado el modelo usando el registro natural de uno de los regresores
```

```python
mod.params
```

```python
mod.mse_resid
```

```python
print (mod.summary())
```
